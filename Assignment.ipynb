```python
###########
# Task 2 ##
###########

#Task 2-a: Find out the race with largest number of records  

#Largest_race_val  = df_data['race'].value_counts().max       37572
Largest_race  = df_data['race'].value_counts().idxmax()
print (">>Task 2-a: The Race with the largest number of records is %s"% (Largest_race))
```

```python
#Task 2-b: #Task 2-b: Find out the total number of doctorate who are married 

num_doctorate = df_data[(df_data['education']=='Doctorate') & (df_data['marital-status']=='married')]
print ("\n\n>>Task 2-b: The total number of doctorate who are married is :- %s"% (num_doctorate))

```

```python
#Task 2-c: Find out the top 10 countries with the highest income.

n=10

top10_countries = df_data[df_data['income > 50K']=='Yes'].sort_values('native-country',ascending=False).groupby("native-country")['hours-per-week'].nunique().nlargest(n)
print("\n\n>>Task 2-c: top 10 countries with the highest income: \n%s" % (top10_countries))


#Task 2-c: Task 2-c: Find out the top 10 counties with the most male.

top10_males= df_data[df_data['gender']=='Male'].sort_values('native-country', ascending=False).groupby('native-country')["age"].nunique().nlargest(n)
print ("\n\n>>Task 2-c: top 10 counties with the most male \n%s" % (top10_males))
```

```python
###########
# Task 3 ##
###########

# Task 3-a: Plot the declaration count for each rating . 
# Think of a way to nicely visualize all the ratings! 
#########################begin code for Task 3-a


countries = set(df_data['native-country'])
print(countries)

race = set(df_data['race'])
print(race)

race_counts = df_data[df_data.race.isin(['Black', 'White', 'Other', 'Asian-Pac-Islander', 'Amer-Indian-Eskimo'])]
```

```python
race_counts.shape
```

```python
sns.set(font_scale = 1.3)
sns.set_style('darkgrid')
plt.figure(figsize=(100,100))
sns.catplot(x="race", data=race_counts ,kind="count",height=7, aspect=1.5,legend=True)
plt.title("Declaration Count for each Rating")
plt.xlabel("Race", size=20)
plt.ylabel("Hours-per-week", size=20);
#########################end code for Task 3-a
```

```python
# Task 3-b: Draw a pie chart that represents native country

#########################begin code for Task 3-b 


country_count = df_data['native-country'].value_counts().to_dict()
print(country_count)
```

```python
labels = []
data = []
for i in country_count:
    labels.append(i)
    data.append(country_count[i])
    
print(labels)
print(data)
```

```python
colors = sns.color_palette('pastel')[0:5]
plt.pie(data, labels = labels, colors = colors,autopct = "%0.2f%%",radius = 3.5)
plt.show()

#########################end code for Task 3-b
```
